```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "resources/schools_complete.csv"
student_data_to_load = "resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_data_to_load)
student_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
main_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
main_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#school & budget calculations using school-only data
dist_schools = len(school_df["school_name"])
dist_budget = school_df["budget"].sum()

#student & grade calculations using main dataframe
dist_students = len(main_df["student_name"])
dist_avg_math = main_df["math_score"].mean()
dist_avg_read = main_df["reading_score"].mean()

#conditional locate fxns to calculate pass rates
dist_pass_math = main_df.loc[main_df["math_score"] >= 70]["student_name"].count() / dist_students
dist_pass_read = main_df.loc[main_df["reading_score"] >= 70]["student_name"].count() / dist_students
dist_pass_all = main_df.loc[(main_df["reading_score"] >= 70) & (main_df["math_score"] >= 70)]["student_name"].count() / dist_students

#display in dataframe, reformat values accordingly
dist_summary = {"total schools": [dist_schools],
                "total students": ["{:,}".format (dist_students)],
                "total budget": ["${:,.2f}".format (dist_budget)],
                "average math score": ["{:.2f}".format (dist_avg_math)],
                "average reading score": ["{:.2f}".format (dist_avg_read)],
                "% passing math": ["{:.2%}".format (dist_pass_math)],
                "% passing reading": ["{:.2%}".format (dist_pass_read)],
                "% overall passing": ["{:.2%}".format (dist_pass_all)]}

dist_summary_df = pd.DataFrame (dist_summary)

dist_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#group main dataframe by school
school_grouped = main_df.groupby(["school_name"])
school_students = school_grouped["student_name"].count()

#pull school type & budget from school-only data
school_type = school_df.set_index("school_name")["type"]
school_budget = school_df.set_index("school_name")["budget"]

#basic calculations
budget_per_student = school_budget / school_students
school_avg_math = school_grouped["math_score"].mean()
school_avg_read = school_grouped["reading_score"].mean()

#conditional locate & group fxns to calculate pass rates by school; unable to use the locate fxn on already grouped data
school_pass_math = (main_df.loc[main_df["math_score"] >= 70].groupby(["school_name"])["student_name"].count()) / school_students
school_pass_read = (main_df.loc[main_df["reading_score"] >= 70].groupby(["school_name"])["student_name"].count()) / school_students
school_pass_all = (main_df.loc[(main_df["math_score"] >= 70) & (main_df["reading_score"] >= 70)].groupby(["school_name"])["student_name"].count()) / school_students

#display in dataframe
school_summary_df = pd.DataFrame ({"school type": school_type,
                                   "total students": school_students,
                                   "total budget": school_budget,
                                   "per student budget": budget_per_student,
                                   "average math score": school_avg_math,
                                   "average reading score": school_avg_read,
                                   "% passing math": school_pass_math,
                                   "% passing reading": school_pass_read,
                                   "% overall passing": school_pass_all})

#make copy of summary dataframe for formatting
school_summary_format = school_summary_df.copy()

#reformat values w/ map fxn (rather than style, which creates a non-sortable dataframe)
school_summary_format["total students"] = school_summary_format["total students"].map("{:,}".format)

school_summary_format["total budget"] = school_summary_format["total budget"].map("${:,.2f}".format)
school_summary_format["per student budget"] = school_summary_format["per student budget"].map("${:,.2f}".format)

school_summary_format["average math score"] = school_summary_format["average math score"].map("{:.2f}".format)
school_summary_format["average reading score"] = school_summary_format["average reading score"].map("{:.2f}".format)

school_summary_format["% passing math"] = school_summary_format["% passing math"].map("{:.2%}".format)
school_summary_format["% passing reading"] = school_summary_format["% passing reading"].map("{:.2%}".format)
school_summary_format["% overall passing"] = school_summary_format["% overall passing"].map("{:.2%}".format)

school_summary_format
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#sort values from school summary dataframe, isolate top 5 values
top_schools_df = school_summary_format.sort_values(by = ["% overall passing"], ascending = False)
top_schools_df = top_schools_df[:5]

top_schools_df
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
#sort values from school summary dataframe, isolate bottom 5 values
bottom_schools_df = school_summary_format.sort_values(by = ["% overall passing"])
bottom_schools_df = bottom_schools_df[:5]

bottom_schools_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#set index of original dataframe to school names
school_index_df = main_df.set_index ("school_name")

#separate dataframe by grade
grade_9 = school_index_df.loc [school_index_df["grade"] == "9th", :]
grade_10 = school_index_df.loc [school_index_df["grade"] == "10th", :]
grade_11 = school_index_df.loc [school_index_df["grade"] == "11th", :]
grade_12 = school_index_df.loc [school_index_df["grade"] == "12th", :]

#group students by school, calculate mean math scores, round to 2 decimal places
grade_9_math = round (grade_9.groupby(["school_name"])["math_score"].mean(), 2)
grade_10_math = round (grade_10.groupby(["school_name"])["math_score"].mean(), 2)
grade_11_math = round (grade_11.groupby(["school_name"])["math_score"].mean(), 2)
grade_12_math = round (grade_12.groupby(["school_name"])["math_score"].mean(), 2)

#display in dataframe
grade_math_df = pd.DataFrame ({"9th": grade_9_math,
                               "10th": grade_10_math,
                               "11th": grade_11_math,
                               "12th": grade_12_math})

#get rid of extraneous index name row
grade_math_df.index.name = None

grade_math_df
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#group students by school, calculate mean reading scores, round to 2 decimal places
grade_9_read = round (grade_9.groupby(["school_name"])["reading_score"].mean(), 2)
grade_10_read = round (grade_10.groupby(["school_name"])["reading_score"].mean(), 2)
grade_11_read = round (grade_11.groupby(["school_name"])["reading_score"].mean(), 2)
grade_12_read = round (grade_12.groupby(["school_name"])["reading_score"].mean(), 2)

#display in dataframe
grade_read_df = pd.DataFrame ({"9th": grade_9_read,
                               "10th": grade_10_read,
                               "11th": grade_11_read,
                               "12th": grade_12_read})

#get rid of extraneous index name row
grade_read_df.index.name = None

grade_read_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#create list of bins & corresponding labels
spend_bins = [0, 584, 629, 644, 675]
spend_labels = ["< $585", "$585-629", "$630-644", "$645-675"]

#create copy of school summary dataframe, add column for bins
spend_df = school_summary_df.copy()
spend_df["spending range (per student)"] = pd.cut(spend_df["per student budget"], spend_bins, labels = spend_labels, include_lowest = True)

#group dataframe by spending range bins, calculate averages for select columns
spend_group = spend_df.groupby("spending range (per student)")

spend_stat_df = spend_group[["average math score",
                             "average reading score",
                             "% passing math",
                             "% passing reading",
                             "% overall passing"]].mean()

#reformat values to a new dataframe
spend_stat_format = spend_stat_df.style.format({"average math score": "{:.2f}".format,
                                                "average reading score": "{:.2f}".format,
                                                "% passing math": "{:.2%}".format,
                                                "% passing reading": "{:.2%}".format,
                                                "% overall passing": "{:.2%}".format})

spend_stat_format
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#create list of bins & corresponding labels
size_bins = [0, 999, 1999, 5000]
size_labels = ["small (<1000)", "medium (1000-1999)", "large (2000-5000)"]

#create copy of school summary dataframe, add column for bins
size_df = school_summary_df.copy()
size_df["school size"] = pd.cut(size_df["total students"], size_bins, labels = size_labels, include_lowest = True)

#group dataframe by school size bins, calculate averages for select columns
size_group = size_df.groupby("school size")

size_stat_df = size_group[["average math score",
                           "average reading score",
                           "% passing math",
                           "% passing reading",
                           "% overall passing"]].mean()

#reformat values to a new dataframe
size_stat_format = size_stat_df.style.format({"average math score": "{:.2f}".format,
                                              "average reading score": "{:.2f}".format,
                                              "% passing math": "{:.2%}".format,
                                              "% passing reading": "{:.2%}".format,
                                              "% overall passing": "{:.2%}".format})

size_stat_format
```
