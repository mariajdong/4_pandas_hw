### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
main_df = pd.read_csv(file_to_load)
main_df.head()
```

## Player Count

* Display the total number of players


```python
#filter out unique usernames, calculate length
player_count = len (main_df["SN"].unique())

#display in dataframe
player_count_df = pd.DataFrame ({"total players": [player_count]})
player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#complete calculations from original dataframe
item_count = len(main_df["Item ID"].unique())
avg_price = round(main_df["Price"].mean(), 2)
purchase_count = len(main_df)
total_rev = main_df["Price"].sum()

#display all floats w/ two decimal places as '$'
pd.options.display.float_format = '${:,.2f}'.format

#display calculations in dataframe
purchase_analysis_df = pd.DataFrame ({"# of unique items": [item_count],
                                     "average price": [avg_price],
                                     "# of purchases": [purchase_count],
                                     "total revenue": [total_rev]})
purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#create new dataframe from username & gender columns of original data
user_df = main_df[["SN", "Gender"]]

#drop repeat usernames for users that made more than one purchase
user_df = user_df.drop_duplicates(subset="SN")

#calculate corresponding count and percentage
gender_count = user_df["Gender"].value_counts()
gender_percent = user_df["Gender"].value_counts(normalize=True)

#display in dataframe
gender_demo_df = pd.DataFrame({"total count": gender_count,
                               "% of players": gender_percent})

#format percentage w/ '%'
gender_demo_format = gender_demo_df.style.format({"% of players": "{:.2%}".format})
gender_demo_format
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#group purchase data by gender
gender_group = main_df.groupby(["Gender"])

#complete calculations w/ grouped data
gender_buy_count = gender_group["SN"].count()
gender_avg_price = round(gender_group["Price"].mean(), 2)
gender_total_price = gender_group["Price"].sum()
gender_avg_person = round(gender_total_price / gender_count, 2)

#display results in dataframe
gender_stat_df = pd.DataFrame({"purchase count": gender_buy_count,
                              "average purchase price": gender_avg_price,
                              "total purchase value": gender_total_price,
                              "average total per person": gender_avg_person})

gender_stat_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create list of bins & corresponding labels
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
age_labels = ["<10", "10-14", "14-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#create copy of original dataframe, add column for bins
age_df = main_df
age_df["age group"] = pd.cut(age_df["Age"], age_bins, labels = age_labels, include_lowest = True)

#drop repeat usernames before grouping by age bins
age_unique_df = age_df.drop_duplicates(subset="SN")
age_unique_group = age_unique_df.groupby ("age group")

#calculate corresponding count and percentage; alternative steps to what was done for the gender demographics
age_count = age_unique_group["SN"].count()
age_percent = age_count / player_count

#display in dataframe
age_demo_df = pd.DataFrame ({"total count": age_count,
                             "% of players": age_percent})

#format percentage w/ '%'
age_demo_format = age_demo_df.style.format({"% of players": "{:.2%}".format})
age_demo_format
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create new grouped dataframe with all purchases, including duplicate users
age_group = age_df.groupby ("age group")

#complete calculations w/ grouped data
age_buy_count = age_group["SN"].count()
age_avg_price = round(age_group["Price"].mean(), 2)
age_total_price = age_group["Price"].sum()
age_avg_person = round(age_total_price / age_count, 2)

#display results in dataframe
age_stat_df = pd.DataFrame({"purchase count": age_buy_count,
                            "average purchase price": age_avg_price,
                            "total purchase value": age_total_price,
                            "average total per person": age_avg_person})

age_stat_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group original purchase data by username
user_group = main_df.groupby ("SN")

#complete calculations w/ grouped data
user_buy_count = user_group["Purchase ID"].count()
user_avg_price = round(user_group["Price"].mean(), 2)
user_total_price = user_group["Price"].sum()

#display results in dataframe
user_stat_df = pd.DataFrame ({"purchase count": user_buy_count,
                              "average purchase price": user_avg_price,
                              "total purchase value": user_total_price,})

#sort by descending total purchase value
user_sort_df = user_stat_df.sort_values (by = ["total purchase value"], ascending = False)

user_sort_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group original purchase data by item ID and item name
item_group = main_df.groupby (["Item ID","Item Name"])

#complete calculations w/ grouped data
item_count = item_group["Purchase ID"].count()
item_price = item_group["Price"].mean()
item_total_price = item_group["Price"].sum()

#display results in dataframe
item_stat_df = pd.DataFrame ({"purchase count": item_count,
                              "item price": item_price,
                              "total purchase value": item_total_price,})

#sort by descending purchase quantity
popular_item_df = item_stat_df.sort_values (by = ["purchase count"], ascending = False)

popular_item_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#sort item statistics dataframe by descending total purchase value
profit_item_df = item_stat_df.sort_values (by = ["total purchase value"], ascending = False)

profit_item_df.head()
```
